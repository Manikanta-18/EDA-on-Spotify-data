# Preprocessing

```python
df.columns
```

```python
columns_to_convert = [ 'in_spotify_playlists','in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts',
       'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm', 'danceability_%', 'valence_%', 'energy_%',
       'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']

for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column], errors='coerce')#converting all the non integer values into Nan
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True) #dropping all the null values
```

```python
df.shape
```

```python
df.dtypes
```

```python
columns_to_int = [ 'in_spotify_playlists','in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts',
       'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm', 'danceability_%', 'valence_%', 'energy_%',
       'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']

for column in columns_to_int:
    df[column] =df[column].astype(int) #converting all the float numeric columns to integer
```

```python
df.info()
```

```python
df.describe()
```

```python
#Adding the values in all playlists
df['total_playlists']=df['in_spotify_playlists']+df['in_apple_playlists']+df['in_deezer_playlists']
```

```python
#Adding the values in all charts
df['total_charts']=df['in_spotify_charts']+df['in_apple_charts']+df['in_deezer_charts']+df['in_shazam_charts']
```

```python
df
```

```python
#dropping the unnecessary columns
df1=df.drop(columns=['artist_count', 'released_month', 'released_day', 'in_spotify_playlists',
                 'in_spotify_charts', 'in_apple_playlists', 'in_apple_charts',
                 'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts'])
```
